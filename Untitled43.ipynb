```python
df = pd.read_csv('Churn_Modelling (1).csv')
```

```python
df.head(2)
```

```python
df.drop('Surname', axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df['Geography'].value_counts()
```

```python
def print_unique_col_values(df):
  for columns in df:
    if df[columns].dtypes == 'object':
      print(f"{columns}: {df[columns].unique()}")
```

```python
print_unique_col_values(df)
```

```python
df['Gender'].replace({'Female': 1, 'Male': 0}, inplace=True)
```

```python
df.head(2)
```

```python
df1 = pd.get_dummies(df, columns=['Geography'], dtype='int64')
```

```python
df1.head(2)
```

```python
df1.dtypes
```

```python
col_for_scale = ['CustomerId', 'CreditScore', 'Age', 'Balance', 'EstimatedSalary' ]
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df1[col_for_scale] = sc.fit_transform(df1[col_for_scale])
```

```python
df1.head(2)
```

```python
for col in df1:
  print(f'{col}: {df1[col].unique()}')
```

```python
X = df1.drop('Exited', axis=1)
y = df1['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(14,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred = []
for elements in yp:
  if elements > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```
